## User-defined If Statements

```python
# Create a variable that buckets prices in low, mid, and high ranges
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function
df['price_range']=df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

## If Statements with loc() function

```python
# Create same price range variable as above using loc
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Can now repeat process on full df since it is faster
ords_prods_merged.loc[ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

## If Statements with For-Loops

```python
# Create variable identifying busiest/slowest days of the week for orders. 0=Saturday, busiest day. 4=Wednesday, slowest day
ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
result = []

for value in ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
ords_prods_merged['busiest_day'] = result
```

```python
ords_prods_merged['busiest_day'].value_counts(dropna = False)
```
