## Steps 2-3.

```python
# 2. Find the aggregated mean of the order_id column grouped by department_id
ords_prods_merged.groupby('department_id').agg({'order_number': ['mean']})
```

## Step 4

```python
# 4. Creating loyalty flag using transform and loc functions
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merged.head(15)
```

```python
# Create flag that assigns loyalty label
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merged['loyalty_flag'].value_counts(dropna=False)
```

```python
ords_prods_merged[['user_id', 'loyalty_flag', 'max_order']].head(60)
```

## Step 5

```python
# 5. Determine if product prices vary by loyalty
ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Prices vary slightly by loyalty status but it seems the opposite of what you'd expect. New customers pay the highest price,
on average, then regular customers, then loyal pay the lowest.

## Step 6

```python
# 6. Create a spending flag for each user based on average price across all orders
ords_prods_merged['average_spend'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merged[['user_id', 'average_spend', 'prices']].head(60)
```

```python
ords_prods_merged.groupby('user_id').agg({'prices': ['mean', 'min', 'max']}).head(15)
```

```python
ords_prods_merged.loc[ords_prods_merged['average_spend'] < 10, 'spend_flag'] = 'Low spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_spend'] >= 10, 'spend_flag'] = 'High spender'
```

```python
pd.options.display.max_rows = None
```

```python
ords_prods_merged[['user_id', 'average_spend', 'spend_flag']].head(100)
```

```python
ords_prods_merged['spend_flag'].value_counts(dropna = False)
```

## Step 7

```python
# 7. Create order frequency flag based on days_since_prior_order
ords_prods_merged.groupby('user_id').agg({'days_since_prior_order': ['median']}).head(15)
```

```python
ords_prods_merged['median_order_frequency'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merged[['user_id', 'median_order_frequency', 'days_since_prior_order']].head(15)
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order_frequency'] > 20, 'order_freq_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['median_order_frequency'] > 10) & (ords_prods_merged['median_order_frequency'] <= 20), 'order_freq_flag'] = 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['median_order_frequency'] <=10, 'order_freq_flag'] = 'Frequent customer'
```

```python
ords_prods_merged[['user_id', 'median_order_frequency', 'order_freq_flag']].head(100)
```
