# Step 2. Creating Charts from Chapter

```python
# Bar chart
bar = ords_prods_cust_merged['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Exporting chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Histogram
hist = ords_prods_cust_merged['prices'].plot.hist(bins = 25)
```

```python
# Exporting histogram
hist.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'hist_prices.png'))
```

```python
# Scatterplot
scatterplot = sns.scatterplot(x = 'prices', y = 'prices',data = ords_prods_cust_merged)
```

```python
# Exporting scatterplot
scatterplot.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'scatter_prices.png'))
```

```python
# Line chart
line = sns.lineplot(data = ords_prods_cust_merged, x = 'orders_day_of_week',y = 'prices')
```

```python
# Exporting line chart
line.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'line_prices_dow.png'))
```
