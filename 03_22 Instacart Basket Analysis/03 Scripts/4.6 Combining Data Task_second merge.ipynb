## 2. Import data

```python
# Remember project folder file path
path = r'C:\Users\rstev\Documents\03_22 Instacart Basket Analysis'
```

```python
# Import merged orders_products_combined dataframe
df_ords_prods = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined.pkl'))
```

```python
# Import products dataframe to merge
df_prods = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'products_checked.csv'))
```

```python
# Check shape of products df
df_prods.shape
```

```python
df_prods.head()
```

```python
# Dropping unnecessary columns from merged df
df_ords_prods = df_ords_prods.drop(columns = ['Unnamed: 0','Unnamed: 0.1','_merge'])
```

```python
# Dropping unnecessary columns from prods df
df_prods = df_prods.drop(columns = ['Unnamed: 0'])
```

```python
df_ords_prods.head()
```

```python
df_prods.head()
```

## 3. Merge Data

```python
# I will use an inner merge since the two dataframes match on a product_id variable
df_ords_prods_merged = df_ords_prods.merge(df_prods, on = ['product_id'], indicator = True)
```

```python
df_ords_prods_merged.head()
```

```python
# Confirm merge using merge flag
df_ords_prods_merged['_merge'].value_counts()
```

## 4. Export as pkl file
